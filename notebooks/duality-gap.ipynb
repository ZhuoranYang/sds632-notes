

**Primal Problem (Nonconvex)**

\begin{align*}
\text{Minimize} \quad & f(x) = x^4 - 4x^2 \\
\text{Subject to} \quad & x \leq 1
\end{align*}


**Dual Problem (Simplified)**

\begin{align*}
\text{Maximize} \quad & g(\lambda) \\
\text{Where} \quad & g(\lambda) = \inf_{x} \{x^4 - 4x^2 + \lambda ( x-1) \}, \, \lambda \in \mathbb{R}
\end{align*}


The primal problem is a straightforward nonconvex optimization problem with a single variable $ x $. The dual problem, as presented, involves finding the maximum of $ g(\lambda) $, which is defined as the infimum of the Lagrangian $ L(x, \lambda) = x^4 - 4x^2 + \lambda (x-1) $ over all $x$ for a given $ \lambda $.

####Observation:
- Dual problem is a convex optimization (maximizing a concave function)
- Duality gap is nonzero for this example

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the primal and dual objective functions
def primal_objective(x):
    return x**4 - 4*x**2

def dual_objective(lam):
    # For simplicity, we're calculating the infimum of the Lagrangian numerically
    x_vals = np.linspace(-10, 10, 400)
    lagrangian_values = [x**4 - 4*x**2 + lam*(x-1) for x in x_vals]
    return min(lagrangian_values)

# Generating data for plotting
x_range = np.linspace(-2, 2, 400)
primal_values = [primal_objective(x) for x in x_range]

lambda_range = np.linspace(0, 5, 400)
dual_values = [dual_objective(lam) for lam in lambda_range]

# Plotting the primal and dual objective functions
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_range, primal_values, label='Primal Objective')
plt.title('Primal Objective Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(lambda_range, dual_values, label='Dual Objective', color='orange')
plt.title('Dual Objective Function')
plt.xlabel('lambda')
plt.ylabel('g(lambda)')
plt.grid(True)

plt.tight_layout()
plt.show()

```

For this example, let's consider a convex optimization problem with an inequality constraint and demonstrate that the duality gap is zero. We'll use a simple quadratic function as the objective function.

**Primal Problem (Convex)**

\begin{align*}
\text{Minimize} \quad & f(x) = x^2 \\
\text{Subject to} \quad & x \geq  1 \\
                         & x \in \mathbb{R}
\end{align*}


To formulate the dual problem, we introduce a non-negative Lagrange multiplier $ \lambda \geq 0 $ for the inequality constraint. The Lagrangian   for this problem is:
$$
L(x, \lambda) = x^2 + \lambda (1 - x)
$$


The dual problem is
\begin{align*}
\text{Maximize} \quad & \lambda - \frac{\lambda^2}{4} \\
\text{Subject to} \quad & \lambda \geq 0
\end{align*}

####Observation:
- Duality gap is zero

```python
# Redefining the primal and dual objective functions for the new problem
def convex_primal_objective(x):
    return x**2

def convex_dual_objective(lam):
    # Dual objective function derived from the Lagrangian
    return lam - lam**2 / 4

# Generating data for plotting
x_range_convex = np.linspace(1, 3, 400)
primal_values_convex = [convex_primal_objective(x) for x in x_range_convex]

lambda_range_convex = np.linspace(0, 4, 400)
dual_values_convex = [convex_dual_objective(lam) for lam in lambda_range_convex]

# Plotting the primal and dual objective functions
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_range_convex, primal_values_convex, label='Primal Objective')
plt.title('Convex Primal Objective Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axvline(x=1, color='red', linestyle='--', label='Constraint (x <= 1)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(lambda_range_convex, dual_values_convex, label='Dual Objective', color='orange')
plt.title('Convex Dual Objective Function')
plt.xlabel('lambda')
plt.ylabel('g(lambda)')
plt.grid(True)

plt.tight_layout()
plt.show()

```
